# Nearest Neighbors Chains Algorithm for k-attribute Narcissistic Stable Matching Problem

k-attribute : every candidate has k characteristics 

narcissistic : k characteristics which have different importance for each candidate 

symmetric: find the dot product of respective attributes and not in the given order

```python
import random
from random import randrange
```

```python
n = int(input("Insert number of men/women: "))
k = int(input("Insert k: "))

# attributes names
attr = [None] * k
for i in range(len(attr)):
    x = "attr" + str(i)
    attr[i] = x
mnames = ['Tom', 'Tim', 'Mike', 'Aaron', 'Abe', 'Acer', 'Ryan', 'Sam', 'Scott']
wnames = ['Abi', 'Aliz', 'Chloe', 'Ella', 'Gwen', 'Isla', 'Kira', 'Lisa', 'Tess']

men = {}
women = {}

for y in range(n):
    d = {}
    for i in range(k):
        a = random.choice(attr)
        while a in d:
            a = random.choice(attr)
        key = a
        d[key] = randrange(1,10)
    key = mnames[y] #select men name
    men[key] = d
    
for y in range(n):
    d = {}
    for i in range(k):
        a = random.choice(attr)
        while a in d:
            a = random.choice(attr)
        key = a
        d[key] = randrange(1,10)
    key = wnames[y] #select men name
    women[key] = d
    
print(men)
print("\n---------------------\n")
print(women)
```

```python
def find_dotP(x,y):
    p = 0
    for key, value in x.items():
        p = p + value * y[key]
    return p
```

```python
men2 = {}
women2 = {}

for key1, value1 in men.items():
    x = men[key1]
    temp = []
    for key2, value2 in women.items():
        y = women[key2]
        product = find_dotP(x,y)
        print(key1, key2, product)
        temp.append((key2,product))
        temp.sort(key=lambda tup: tup[1], reverse=True)
    l = []
    for i in range(len(temp)):
        l.append(temp[i][0])
    print(l)
    men2[key1] = l
    print("\n - - - - - - - - - - - - - \n")

for key1, value1 in women.items():
    x = women[key1]
    temp = []
    for key2, value2 in men.items():
        y = men[key2]
        product = find_dotP(x,y)
        print(key1, key2, product)
        temp.append((key2,product))
        temp.sort(key=lambda tup: tup[1], reverse=True)
    l = []
    for i in range(len(temp)):
        l.append(temp[i][0])
    print(l)
    women2[key1] = l
    print("\n - - - - - - - - - - - - - \n")    

print(men2)
print("\n---------------------\n")
print(women2)
```

```python
# traditional stable matching

def stable_marriage(men_dict, wom_dict):
    
    # Initialize dictionary of final matching.
    pairing_list = {}
    free_men_list = []
    
    # Convert dictionary of proposers into list.
    for key, value in wom_dict.items():
        pairing_list[key] = None
    
    for key, value in men_dict.items():
        free_men_list.append(key)

    # Iterates through list of proposers. 
    # Stops when every proposer is in a marriage.
    while free_men_list != []: 

        #Each proposer proposes to their preferred proposee.
        proposer = free_men_list[0]
        proposee = men_dict[proposer][0] 
        proposee_pref = wom_dict[proposee].index
        current_suitor = pairing_list[proposee]
            
        # If proposee is unmatched or is in a less stable matching, the proposer is matched with the proposee.
        if current_suitor == None:
        
            free_men_list.remove(proposer)
            pairing_list[proposee] = proposer
            
            print(proposer +' proposes and marries '+ proposee + '.')
        
        elif proposee_pref(current_suitor) > proposee_pref(proposer):
        
            free_men_list.append(current_suitor)
            free_men_list.remove(proposer)
            men_dict[current_suitor].remove(proposee)
            pairing_list[proposee] = proposer
            
            print(proposer + ' intercepts ' + current_suitor + ' and marries ' + proposee + '.')
        
        else:
            
            men_dict[proposer].remove(proposee)
            print(proposer + ' proposes and is rejected by ' + proposee + '.')
            
    print("\n----Happily ever after -----")
    print(pairing_list)
    
stable_marriage(men2, women2)   
```

```python
# Nearest Neighbor Chains

def NNC_SM(men_dict, wom_dict):
    
    # Initialize the matching empty
    free_men_list = []
    pairs = {}    
    
    # convert men dict to list
    for key, value in men_dict.items():
        free_men_list.append(key)

    # Initialize an empty stack S
    S = []
    
    # Repeat while there is an unmatched men
    while free_men_list != []:
        
        # If S is empty, add any unmatched men to it.
        if len(S) == 0:
            S.append(free_men_list[randrange(len(free_men_list))])
            
        # Let p be the agent at the top of the stack
        p = S[-1]
        print("p = ", p)
        
        # find p's nearest neighbor q of the opposite set
        if p in men_dict:
            q = men_dict[p][0]
            flag = 1 # counter for preference list
            print("q = ", q)
            while flag > 0:
                if len(pairs) != 0:
                    if q in pairs.values():
                        for key, value in pairs.items():
                            if q == value:
                                print("already matched with ", key)
                                # q is already matched
                                #keep the most prefered option
                                lover = wom_dict[q] #list
                                if lover.index(key) > lover.index(p):
                                    print("destroy marriage")
                                    #destroy previous marriage 
                                    del pairs[key]
                                    # ex husband becomes free man
                                    free_men_list.append(key)
                                    flag = -1
                                    break;
                                else:
                                    # q is already married and prefers current husband move to next choice
                                    q = men_dict[p][flag]
                                    flag = flag + 1
                                    print("new q = ", q)
                    else:
                        break;
                else:
                    break;
        else:
            q = wom_dict[p][0]
            flag = 1 # counter for preference list
            print("q = ", q)
            while flag > 0:
                if len(pairs) != 0:
                    if q in pairs:
                        value = pairs[q]
                        print("already matched with ", value)
                        # q is already matched
                        #keep the most prefered option
                        lover = men_dict[q] #list
                        if lover.index(value) > lover.index(p):
                            print("destroy marriage")
                            #destroy previous marriage
                            del pairs[key]
                            flag = -1
                            break;
                        else:
                            # q is already much and prefers current wife move to next choice
                            q = wom_dict[p][flag]
                            flag = flag + 1
                            print("new q = ", q)
                    else:
                        break;
                else:
                    break;
                        
        print("nn = ", q)
        print("pairs: ", pairs)
        
        # If q is not already in S, add it
        if q not in S:
            S.append(q)
            print("Just added q = ", q, " in S")
        else:
            # Otherwise, q must be the second-from-top element in S
            if q == S[-2]:
#                 print("q=", q," must be the second-from-top element in S")
                a = S.pop()
                b = S.pop()
                print("Marry: ",a,b)
                if a in free_men_list:
                    pairs[a] = b
                else:
                    pairs[b] = a
                    
                if a in free_men_list:
                    free_men_list.remove(a)
                    print(free_men_list)
                elif b in free_men_list:
                    free_men_list.remove(b)
                    print(free_men_list)
    print("Happily Ever After")
    print(pairs)
    
NNC_SM(men2, women2)
```
